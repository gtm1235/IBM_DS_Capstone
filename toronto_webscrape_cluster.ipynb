<h4>***Import necessary modules for data cleaning and manipulation -- in this case Pandas***</h4>

```python
import pandas as pd
```

<h4>Using data from Wikipedia I will import the table into Pandas, format it, and ,view to see if loaded correctly<h4>

```python
url = r'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
```

```python
table_toronto = pd.read_html(url)
table_toronto = table_toronto[0]
table_toronto.head()
```

<h4>I will now first remove any rows where the Borough is not assigned and reset index</h4>

```python
pd_borough= table_toronto.loc[table_toronto['Borough'] != 'Not assigned',].reset_index()
pd_borough = pd_borough[['Postcode', 'Borough','Neighbourhood']]
pd_borough.head()
```

<h4>I am creating a pandas groupby object to stack postal codes for easier manipulation later</h4>

```python
df_by_pc = pd_borough.groupby('Postcode')
df_by_pc.head(5)
```

<h4>Below I am creating an empty data freame for later use</h4>

```python
nb_df_combine = pd.DataFrame(columns = list(pd_borough.columns))
nb_df_combine
```

<h4>     Here I iterate through a list of unique postal codes and append the neighborhoods in each 
     to a list. I then append this list to the newly empty <br> <br>datarame created earlier along with 
     the correct Postal code and Borough name. I made the assumption that specific Boroughs were 
    in Specific Postal <br><br> codes and verified by viewing the data</h4>

```python
temp_list = []
#find number of unique postcodes from grouby dataset. I use list to access information from the groupby object
for i in range(len(df_by_pc['Postcode'].unique())):
    for nbh in list(df_by_pc)[i][1]['Neighbourhood']:
         temp_list.append(nbh) 
    #combine list of neighborhoods into a comma seperated string        
    temp_list_1 = ', '.join(str(e) for e in temp_list)
    #Create row for appending to the dataset
    data = {'Neighbourhood': temp_list_1,
            'Postcode':(list(df_by_pc)[i][1]['Postcode'].unique())[0],
           'Borough': pd_borough[pd_borough['Postcode']== (list(df_by_pc)[i][1]['Postcode'].
                                                           unique())[0]]['Borough'].unique()[0]}
    nb_df_combine = nb_df_combine.append(data, ignore_index=True)
    temp_list = []
nb_df_combine.head()
```

<h4> here I rename "Not assigned" Neighbourhoods with Borough names</h4>

```python
nb_df_combine.loc[nb_df_combine['Neighbourhood']=='Not assigned', ['Neighbourhood']] = nb_df_combine['Borough']
nb_df_combine.head(25)
```

<h4> Install geocoder library and use to obtain Long and Lat for postal codes. Adding while loop to be sure values are available</h4>

```python
import geocoder # import geocoder

for row in nb_df_combine.iterrows():
    postal_code = row[1][0]
    # initialize your variable to None
    lat_lng_coords = None

    # loop until you get the coordinates
    while(lat_lng_coords is None):
      g = geocoder.google('{}, Toronto, Ontario'.format(postal_code))
      lat_lng_coords = g.latlng
      print(lat_lng_coords)

    latitude = lat_lng_coords[0]
    longitude = lat_lng_coords[1]
    print(postal_code, latitude, longitude)
    nb_df_combine['Latitude']=latitude
    nb_df_combine['Longitude']=longitude
```
