```python
import pandas as pd
```

```python
url = r'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
```

```python
table_toronto = pd.read_html(url)
```

```python
table_toronto = table_toronto[0]
```

```python
table_toronto
```

```python
pd_borough= table_toronto.loc[table_toronto['Borough'] != 'Not assigned',].reset_index()
pd_borough = pd_borough[['Postcode', 'Borough','Neighbourhood']]
pd_borough.head()
```
